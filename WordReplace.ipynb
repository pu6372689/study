```python
import pandas as pd
import os
import re
import jieba
```

```python
#確認資料
df = pd.read_csv('all_ver4.csv')
df
```

```python
#生成過濾用的字典
combine_dict = {}

f = open("../same_meaning.txt", "r", encoding='utf-8')
            # same_meaning.txt是同義詞表，用tab分割
for line in f:
    seperate_word = line.strip().split("\t")
    num = len(seperate_word)
    for i in range(1, num):
        
        combine_dict[seprate_word[i]] = seperate_word[0]
        #以此形式儲存 KEY為相似詞 : VALUE為目標詞
f.close()
```

```python
def same_meaning(lenn):
    food = ''
    for i in str(df['ingredients'][lenn]).split(','):
        if i in combine_dict:
            food += combine_dict[i] + ','
        else:
            food += i+','
        #使用if..else..進行篩選
        
    df.loc[lenn,'ingredients'] = food.rstrip(',')
    #直接對dataframe做變更
```

```python
#從第一項執行到最後一項
lenn = 0
while lenn < len(df):
    same_meaning(lenn)
    lenn += 1
```

```python
#儲存
df.to_csv('all_ver4.csv',index = 0, encoding='utf_8_sig')
```
